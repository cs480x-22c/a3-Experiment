```python
import pandas as pd

df = pd.read_csv('experiment_data.csv')
df.drop(df[df['true_percentage'] == 0].index, inplace = True)
```

```python
import numpy as np

df_pie = df[df['Chart_Type'] == "pie"]
df_pie['log2error'] = np.log2(np.abs(df_pie['guessed_percentage']-df_pie['true_percentage'])+(1/8))
df_pie
```

```python
df_pie3D = df[df['Chart_Type']== 'pie3D']
df_pie3D['log2error'] = np.log2(np.abs(df_pie3D['guessed_percentage']-df_pie3D['true_percentage'])+(1/8))
df_pie3D
```

```python
df_stacked3D = df[df['Chart_Type']== 'stackedBar3D']
df_stacked3D['log2error'] = np.log2(np.abs(df_stacked3D['guessed_percentage']-df_stacked3D['true_percentage'])+(1/8))
df_stacked3D
```

```python
df_stacked = df[df['Chart_Type']== 'stackedBar']
df_stacked['log2error'] = np.log2(np.abs(df_stacked['guessed_percentage']-df_stacked['true_percentage'])+(1/8))
df_stacked
```

```python
import scipy.stats as st

data_pie = list(df_pie['log2error'])
data_pie3D = list(df_pie3D['log2error'])
data_stacked3D = list(df_stacked3D['log2error'])
data_stacked = list(df_stacked['log2error'])
```

```python
import matplotlib.pyplot as plt
import statistics
from math import sqrt


def plot_confidence_interval(x, values, z=1.96, color='#2187bb', horizontal_line_width=0.25):
    mean = statistics.mean(values)
    stdev = statistics.stdev(values)
    confidence_interval = z * stdev / sqrt(len(values))

    left = x - horizontal_line_width / 2
    top = mean - confidence_interval
    right = x + horizontal_line_width / 2
    bottom = mean + confidence_interval
    plt.plot([top, bottom],[x, x], color=color)
    plt.plot([top, top], [left, right], color=color)
    plt.plot([bottom, bottom], [left, right], color=color)
    plt.plot(mean, x, 'o', color='#f44336')

    return mean, confidence_interval


plt.yticks([1, 2, 3, 4], ['pie', 'pie3D', 'stacked3D', 'stacked'])

plt.title('Results')
plt.xlabel('Log Error')
plt.ylabel('Chart Type')
plt.xlim([2.0, 3.0])
plot_confidence_interval(1, data_pie)
plot_confidence_interval(2, data_pie3D)
plot_confidence_interval(3, data_stacked3D)
plot_confidence_interval(4, data_stacked)
plt.show()
```
